```python
import json
```

```python
# Choose a cleaned file
file_path = "../data/ZemenExpress_cleaned.json"

# Load the messages
with open(file_path, "r", encoding="utf-8") as f:
    cleaned_messages = json.load(f)

# Select 50 for labeling
subset_to_label = cleaned_messages[:50]

# Show them for review
for i, msg in enumerate(subset_to_label, 1):
    print(f"{i}. {msg['cleaned_text']}\n")
```
